```python
from rdflib import Graph, Namespace, Literal, RDF
import os


# URL del archivo JSON
json_url = "https://deinsoluciones.cl/web_semantica/datos_importaciones.json"

# Descargar el JSON desde la URL
response = requests.get(json_url)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    data = response.json()  # Convertir la respuesta a JSON
else:
    print(f"Error al obtener el archivo JSON. Código de estado: {response.status_code}")
    exit()

# Lista de categorías y su versión singular para asignar los prefijos correctos
categorias = ['aduanas', 'paises', 'puertos']
singular = ['aduana', 'pais', 'puerto']

# Definir namespaces con prefijo vacío para "http://example.org/"
default_ns = Namespace("http://example.org/")
xsd = Namespace("http://www.w3.org/2001/XMLSchema#")
schema = Namespace("http://schema.org/")

# Prefijos deseados para el archivo final
desired_prefixes = """\
prefix :       <http://example.org/>
prefix xsd:    <http://www.w3.org/2001/XMLSchema#>
prefix rdf:    <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
prefix schema: <http://schema.org/>
"""

# Procesar cada categoría
for i, categoria in enumerate(categorias):
    if categoria in data:
        # Extraer la información de la categoría actual
        registros = data[categoria]

        # Convertir la información a un DataFrame
        df_registros = pd.DataFrame(registros)

        # Crear un grafo RDF
        g = Graph()
        g.bind("", default_ns)  # Prefijo vacío
        g.bind("xsd", xsd)
        g.bind("schema", schema)

        # Obtener el nombre en singular correspondiente a la categoría
        singular_categoria = singular[i]

        # Agregar datos al grafo RDF utilizando el valor del campo 'Código'
        for _, row in df_registros.iterrows():
            item_id = f"{singular_categoria}{row['Código']}"  # Usar código del JSON
            item_uri = default_ns[item_id]

            g.add((item_uri, RDF.type, default_ns[singular_categoria.capitalize()]))  # Usa la versión singular capitalizada
            g.add((item_uri, schema.nombre, Literal(row['Glosa'], datatype=xsd.string)))

        # Guardar el grafo en formato RDF temporal
        temp_file = f"{categoria}_temp.ttl"
        g.serialize(destination=temp_file, format="turtle")

        # Leer el archivo generado y modificar su contenido
        with open(temp_file, 'r', encoding='utf-8') as file:
            lines = file.readlines()

        # Procesar líneas para realizar las modificaciones
        processed_lines = []
        for line in lines:
            if not line.startswith("@prefix"):  # Eliminar prefijos generados automáticamente
                line = line.replace(" a ", " rdf:type ")  # Reemplazar ' a ' por ' rdf:type '
                line = line.replace("schema1:", "schema:")  # Corregir prefijo schema
                processed_lines.append(line)

        # Guardar el resultado final con los cambios
        output_file = f"{categoria}.ttl"
        with open(output_file, 'w', encoding='utf-8') as file:
            file.write(desired_prefixes + "\n")
            file.writelines(processed_lines)

        # Eliminar el archivo temporal
        os.remove(temp_file)

        print(f"✅ Archivo RDF generado con éxito: '{output_file}'")
    else:
        print(f"⚠️ Advertencia: La categoría '{categoria}' no está presente en el JSON.")
```

```python
from rdflib import Graph, Namespace, Literal, RDF, XSD
import os

# Cargar el archivo CSV
csv_file = "https://deinsoluciones.cl/web_semantica/df_importadores_unicos.csv"  # Reemplaza con la ruta de tu archivo CSV
df = pd.read_csv(csv_file,delimiter=",",encoding='utf-8')

# Crear un grafo RDF
g = Graph()

# Definir namespaces con prefijo vacío para "http://example.org/"
default_ns = Namespace("http://example.org/")
xsd = Namespace("http://www.w3.org/2001/XMLSchema#")
schema = Namespace("http://schema.org/")

# Vincular namespaces
g.bind("", default_ns)
g.bind("xsd", xsd)
g.bind("schema", schema)

# Agregar datos al grafo RDF utilizando el prefijo vacío
for _, row in df.iterrows():
    cliente_id = f"cliente{row['ID']}"
    cliente_uri = default_ns[cliente_id]

    g.add((cliente_uri, RDF.type, default_ns.Importador))
    g.add((cliente_uri, schema.comuna_importador, Literal(row['NOMBRE_COMUNA'], datatype=XSD.string)))
    # Convertir a string explícitamente para que aparezca con comillas
    g.add((cliente_uri, schema.numero_unico_importador, Literal(str(row['NUM_UNICO_IMPORTADOR']), datatype=XSD.integer)))

# Guardar el grafo en formato RDF temporal
temp_file = "importadores_temp.ttl"
g.serialize(destination=temp_file, format="turtle")

# Leer el archivo generado y modificar su contenido para ajustar el formato
with open(temp_file, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Nueva cabecera con los prefijos correctos
desired_prefixes = """\
prefix :       <http://example.org/>
prefix xsd:    <http://www.w3.org/2001/XMLSchema#>
prefix rdf:    <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
prefix schema: <http://schema.org/>
"""

# Procesar líneas para eliminar las llaves '{ }' y corregir el prefijo schema1 a schema
processed_lines = []
for line in lines:
    if not line.startswith("@prefix"):  # Eliminar prefijos generados automáticamente
        line = line.replace(" a ", " rdf:type ")  # Reemplazar ' a ' por ' rdf:type '
        line = line.replace("schema1:", "schema:")  # Corregir prefijo schema
        line = line.replace("{", "").replace("}", "")  # Eliminar llaves
        processed_lines.append(line)

# Guardar el resultado final con los cambios
output_file = "importadores.ttl"
with open(output_file, 'w', encoding='utf-8') as file:
    file.write(desired_prefixes + "\n")
    file.writelines(processed_lines)

# Eliminar el archivo temporal
os.remove(temp_file)

print(f"Archivo RDF generado con éxito: '{output_file}'")
```

```python
# Crear una función para calcular las métricas por cada NUMENCRIPTADO
def calcular_metricas(numencriptado):
    # Filtrar en df_resumido por el valor actual de NUMENCRIPTADO
    subset = df_resumido[df_resumido['NUMENCRIPTADO'] == numencriptado]

    # Calcular la suma de CANT_MERC y el máximo de NUMITEM
    cant_merc_sum = subset['CANT_MERC'].sum()
    numitem_max = subset['NUMITEM'].max()

    return pd.Series({'CANT_MERC_SUM': cant_merc_sum, 'NUMITEM_MAX': numitem_max})

# Aplicar la función para calcular las métricas en cada NUMENCRIPTADO único de df_example
metricas = df_example['NUMENCRIPTADO'].apply(calcular_metricas)

# Unir las métricas calculadas al DataFrame df_example
df_example = pd.concat([df_example.reset_index(drop=True), metricas.reset_index(drop=True)], axis=1)
```
